MEDIAN

```python
import cv2
import os
import math
import numpy as np
import cv2
folder = "/content/OCT_Erosion/DME"
k=1
for im in os.listdir(folder):
    path = os.path.join(folder,im)
    img_noisy1 = cv2.imread(path,0)
    m, n = img_noisy1.shape
    img_new1 = np.zeros([m, n])
    for i in range(1, m-1):
        for j in range(1, n-1):
            temp = [img_noisy1[i-1, j-1],
                    img_noisy1[i-1, j],
                    img_noisy1[i-1, j + 1],
                    img_noisy1[i, j-1],
                    img_noisy1[i, j],
                    img_noisy1[i, j + 1],
                    img_noisy1[i + 1, j-1],
                    img_noisy1[i + 1, j],
                    img_noisy1[i + 1, j + 1]]
            temp = sorted(temp)
            img_new1[i, j]= temp[4]
    img_new1 = img_new1.astype(np.uint8)
    cv2.imwrite(os.path.join("/content/drive/MyDrive/OCT Project/OCT Code & Dataset/Image Processing/Median on Erosion/DME", 'M_DME'+str(k)+'.jpeg'),img_new1)
    k = k+1
print("success")
```

Alpha 1 Beta 2

```python
import cv2  
import os  
folder = "/content/drive/MyDrive/OCT Project/OCT Code & Dataset/Image Processing/Median on Erosion/NORMAL"  
j=1  
for im in os.listdir(folder):     
    path = os.path.join(folder,im)  
    image = cv2.imread(path)  
      
    alpha = 1.0 # Contrast control (1.0-3.0)  
    beta = 2 # Brightness control (0-100)  
  
    adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)   
      
    cv2.imwrite(os.path.join('/content/drive/MyDrive/OCT Project/OCT Code & Dataset/Image Processing/Alpha Beta on Median Erosion/NORMAL_A', 'NORMAL_Al '+str(j)+'.jpeg'),adjusted)  
    j = j+1

print("success")
```

Gamma 1.1

```python
import cv2
import os
import math
import numpy as np
import cv2
folder = "/content/drive/MyDrive/OCT Project/OCT Code & Dataset/Image Processing/Median on Erosion/CNV"
k=1
for im in os.listdir(folder):
    path = os.path.join(folder,im)
    img = cv2.imread(path,0)
    gammaa = 1.1
    gamma = np.power(img, gammaa)
    img_new1 = gamma.astype(np.uint8)
    cv2.imwrite(os.path.join("/content/drive/MyDrive/OCT Project/OCT Code & Dataset/Image Processing/Gamma on Median on Erosion/CNVG", 'CNV_G'+str(k)+'.jpeg'),img_new1)
    k = k+1
print("success")
```
