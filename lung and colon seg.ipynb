```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read in the image in grayscale
img = cv2.imread("/content/drive/MyDrive/Colon and lung cancer (histopathological)/colonca1.jpeg", cv2.IMREAD_GRAYSCALE)

# Determine which openCV version were using
if cv2.__version__.startswith('2.'):
    detector = cv2.SimpleBlobDetector()
else:
    detector = cv2.SimpleBlobDetector_create()

# Detect the blobs in the image
keypoints = detector.detect(img)
print(len(keypoints))

# Draw detected keypoints as red circles
imgKeyPoints = cv2.drawKeypoints(img, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display found keypoints
cv2_imshow(img)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
img = cv2.imread("/content/drive/MyDrive/Colon and lung cancer (histopathological)/thresh_to_bitwise.png")

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Define a range of colors for the background
lower_white = np.array([0,0,240])
upper_white = np.array([180,30,255])

# Threshold the image to get only the background pixels
mask = cv2.inRange(hsv, lower_white, upper_white)

# Remove the background from the image
res = cv2.bitwise_and(img,img, mask=mask)

# Show the result
cv2_imshow(res)


```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
img = cv2.imread("/content/drive/MyDrive/Colon and lung cancer (histopathological)/laplacian_of_gaussian.jpg", cv2.IMREAD_GRAYSCALE)

# Threshold the image to get a binary mask
ret,thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

# Define the dilation kernel
kernel = np.ones((5,5), np.uint8)

# Perform the dilation operation
dilation = cv2.dilation(thresh, kernel, iterations=1)

# Show the result
cv2_imshow(dilation)
```
